```python
#For Indra Nagar
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.nobroker.in/property/sale/bangalore/Indra%20Nagar?searchParam=W3sibGF0IjoxMy4wNjk4MjgxLCJsb24iOjc3LjU4OTYwNTQ5OTk5OTk5LCJwbGFjZUlkIjoiQ2hJSnowVW1UU0lZcmpzUm1CRFhNMnZXS3pzIiwicGxhY2VOYW1lIjoiSW5kcmEgTmFnYXIifV0=&radius=2.0&city=bangalore&locality=Indra%20Nagar")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup.prettify
```

```python
title=[]
for i in soup.find_all("a",class_="nb__U5JyW"):
    for j in i.find_all("span"):
        title.append(j.text)
title
```

```python
location=[]
for k in soup.find_all("div",class_="nb__1EwQz"):
    location.append(k.text)
len(location)  
```

```python
area=[]
for k in soup.find_all("div",class_="nb__FfHqA"):
    area.append(k.text)
area  
```

```python
dic={"Title of the house":title,"Location":location,"Area":area}
```

```python
import pandas as pd 
```

```python
df=pd.DataFrame(dic)
df
```

```python
price=[]
alldata=[]
for s in soup.find_all('div',class_="font-semi-bold heading-6"):
    alldata.append(s.text)
```

```python
emi=[]
price=[]
flag=0
for i in alldata:
    if flag==0:
        flag=1
    elif flag==1:
        emi.append(i)
        flag=2
    elif flag==2:
        price.append(i)
        flag=0
    
   
        
```
