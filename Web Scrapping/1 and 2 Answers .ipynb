```python
# Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame
```

```python
ind=requests.get("https://www.imdb.com/list/ls009997493/")
ind
```

```python
indcont=BeautifulSoup(ind.content)
indcont
```

```python
indmovies=[]
for e in indcont.find_all("h3",class_="lister-item-header"):
    for f in e.find_all("a"):
        indmovies.append(f.text)
len(indmovies)
```

```python
yearind=[]
for g in indcont.find_all("h3",class_="lister-item-header"):
    for h in g.find_all("span",class_="lister-item-year text-muted unbold"):
        yearind.append(h.text)
len(yearind)
```

```python
indrating=[]
for rat in indcont.find_all("div",class_="ipl-rating-star small"):
    for rat2 in rat.find_all("span",class_="ipl-rating-star__rating"):
        indrating.append(rat2.text)
len(indrating)
```

```python
dicint={"Name of Indian Movies":indmovies,"Rating":indrating,"Year of Release":yearind,}
dicint
```

```python
indf=pd.DataFrame(dicint)
```
